# 2. Pemrosesan Data

### a. Pembersihan data


Penjelasan :

data dilakukan pembersihan
Note : data kosong adalah Null atau NaN

**Kolom Numerik**

1. Missing Values : Diganti dengan Mean
2. outlier : Diganti dengan Median

**Kolom Kategorical**


1.   Missing values : Diganti dengan nilai yang sering muncul




```python
import pandas as pd

# Pisahkan kolom numerik dan kategorikal setelah menghapus kolom 'ID'
df_cleaned = df.drop(columns=['ID'])

# Pisahkan kolom numerik dan kategorikal
numerical_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns

# Mengganti missing values dengan mean untuk kolom numerik
df_cleaned[numerical_columns] = df_cleaned[numerical_columns].fillna(df_cleaned[numerical_columns].mean())

# Mengganti outlier dengan median untuk kolom numerik menggunakan metode IQR
Q1 = df_cleaned[numerical_columns].quantile(0.25)
Q3 = df_cleaned[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas untuk setiap kolom numerik
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengganti nilai yang dianggap outlier dengan median
for col in numerical_columns:
    is_outlier = (df_cleaned[col] < lower_bound[col]) | (df_cleaned[col] > upper_bound[col])
    df_cleaned.loc[is_outlier, col] = df_cleaned[col].median()

# Mengganti missing values dengan modus (nilai yang paling sering muncul) hanya pada kolom kategorikal
df_cleaned[categorical_columns] = df_cleaned[categorical_columns].apply(lambda col: col.fillna(col.mode()[0]))

# Cek statistik data setelah pembersihan
print("Preview data yang dibersihkan:")
print(df_cleaned.head())

# Tampilkan tabel interaktif dengan data yang sudah dibersihkan
import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(
    header=dict(
        values=list(df_cleaned.columns),
        fill_color='lightblue',
        align='left',
        font=dict(color='black', size=12)
    ),
    cells=dict(
        values=[df_cleaned[col] for col in df_cleaned.columns],
        fill_color='white',
        align='left',
        font=dict(color='black', size=11)
    )
)])

fig.update_layout(
    width=1000,
    height=600,
    title="Tabel Interaktif: Data Cirrhosis (Cleaned with Mean and Median)"
)

fig.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np

# Inisialisasi LabelEncoder untuk mengonversi label kategorikal menjadi numerik
label_encoder = LabelEncoder()

# Konversi y_test menjadi label numerik
y_test_numeric = label_encoder.fit_transform(y_test)

# Mendapatkan probabilitas prediksi dari model (untuk AUC)
y_pred_proba = model.predict_proba(X_test)  # Mendapatkan probabilitas untuk setiap kelas

# Menghitung AUC untuk klasifikasi multikelas
auc = roc_auc_score(y_test_numeric, y_pred_proba, multi_class='ovr')
print(f"AUC: {auc}")

# Visualisasi ROC curve untuk setiap kelas
n_classes = y_pred_proba.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_numeric, y_pred_proba[:, i], pos_label=i)
    roc_auc[i] = auc  # AUC untuk setiap kelas

# Plot ROC curve untuk setiap kelas
plt.figure()

colors = ['blue', 'green', 'red']  # Sesuaikan dengan jumlah kelas Anda
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
